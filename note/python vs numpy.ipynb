```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# CSV 파일 로드
file_path_numpy = "numpy 860.csv"  # 실제 파일 경로로 수정 필요
file_path_python = "python 860.csv"  # 실제 파일 경로로 수정 필요

df_numpy = pd.read_csv(file_path_numpy)
df_python = pd.read_csv(file_path_python)

# 날짜 형식 변환
df_numpy["DateTime"] = pd.to_datetime(df_numpy["DateTime"])
df_python["DateTime"] = pd.to_datetime(df_python["DateTime"])

# 파일명에서 확장자 제거하여 타이틀로 사용
title_numpy = "numpy 860"
title_python = "python 860"

# 서브플롯 설정 (2개 그래프)
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=False)  # 공유 Y축 비활성화

# 첫 번째 그래프 (Python)
ax1 = axes[0]
ax1.plot(df_python["DateTime"], df_python["Mean_Test_Time_(ms)"], color="blue", label="Mean_Test_Time(ms)")
ax1.plot(df_python["DateTime"], df_python["TPS"], color="red", label="TPS")
ax1.set_title(title_python)
ax1.set_xlabel("Time (DateTime)")
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))
ax1.tick_params(axis="x", rotation=45)
ax1.legend()
ax1.grid(True)

# 두 번째 Y축 추가 (MTTFB 확대 표현)
ax1_twin = ax1.twinx()
ax1_twin.plot(df_python["DateTime"], df_python["Mean_time_to_first_byte"], color="gold", linestyle="dashed", label="MTTFB")
ax1_twin.set_ylim(0, 50)  # MTTFB 값 확대
ax1_twin.set_ylabel("MTTFB (ms)", color="gold")
ax1_twin.tick_params(axis="y", labelcolor="gold")

# 두 번째 그래프 (Numpy)
ax2 = axes[1]
ax2.plot(df_numpy["DateTime"], df_numpy["Mean_Test_Time_(ms)"], color="blue", label="Mean_Test_Time(ms)")
ax2.plot(df_numpy["DateTime"], df_numpy["TPS"], color="red", label="TPS")
ax2.set_title(title_numpy)
ax2.set_xlabel("Time (DateTime)")
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))
ax2.tick_params(axis="x", rotation=45)
ax2.legend()
ax2.grid(True)

# 두 번째 Y축 추가 (MTTFB 확대 표현)
ax2_twin = ax2.twinx()
ax2_twin.plot(df_numpy["DateTime"], df_numpy["Mean_time_to_first_byte"], color="gold", linestyle="dashed", label="MTTFB")
ax2_twin.set_ylim(0, 50)  # MTTFB 값 확대
ax2_twin.set_ylabel("MTTFB (ms)", color="gold")
ax2_twin.tick_params(axis="y", labelcolor="gold")

# 전체 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
