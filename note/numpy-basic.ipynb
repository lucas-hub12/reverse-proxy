```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# NumPy 기반 함수
def gen_r_array_choices_numpy(N):
    a = np.random.choice(range(101), size=N)
    b = np.random.choice(range(101), size=N)
    return a, b

# Python random 기반 함수
def gen_r_array_choices(N):
    a = random.choices(range(101), k=N)
    b = random.choices(range(101), k=N)
    return a, b

# # 고차 함수: 실행 시간 측정
# def measure_time(func, N):
#     start = time.time()
#     func(N)
#     return time.time() - start

# 실행 시간 측정 (10번 반복 후 평균값 반환)
def measure_time(func, N, repetitions=10):
    times = []
    for _ in range(repetitions):
        start = time.perf_counter()
        func(N)
        times.append(time.perf_counter() - start)
    return sum(times) / repetitions  # 평균 실행 시간 반환
    
# 테스트할 N 값 생성 (1부터 1,000,000까지 100개 간격)
N_values = np.linspace(1, 3_000, 100, dtype=int)

numpy_times = []
random_times = []
# TODO 각 함수의 실행 시간 계산 위 배열에 추가
for N in N_values:
    numpy_times.append(measure_time(gen_r_array_choices_numpy, N))
    random_times.append(measure_time(gen_r_array_choices, N))

# 선그래프 그리기
plt.figure(figsize=(10, 6)) # 차트 크기 -> 그래프의 가로 크기 10인치, 세로 크기 6인치
plt.xlabel('Array Size') # x축 레이블
plt.ylabel('Time') # y축 레이블
plt.title('Performance : numpy VS random')
plt.grid(True) # 그래프에 그리드 ON
plt.plot(N_values, numpy_times, label='numpy', color='blue')
plt.plot(N_values, random_times, label='random', color='red')
plt.legend() # 범례를 추가
plt.show()
```

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# NumPy randint 기반 함수
def gen_r_array_choices_numpy_rdi(N):
    a = np.random.randint(0, 101, size=N)
    b = np.random.randint(0, 101, size=N)
    return a, b
    
# NumPy 기반 함수
def gen_r_array_choices_numpy(N):
    a = np.random.choice(range(101), size=N)
    b = np.random.choice(range(101), size=N)
    return a, b

# Python random 기반 함수
def gen_r_array_choices(N):
    a = random.choices(range(101), k=N)
    b = random.choices(range(101), k=N)
    return a, b

# # 고차 함수: 실행 시간 측정
# def measure_time(func, N):
#     start = time.time()
#     func(N)
#     return time.time() - start

# 실행 시간 측정 (10번 반복 후 평균값 반환)
def measure_time(func, N, repetitions=10):
    times = []
    for _ in range(repetitions):
        start = time.perf_counter()
        func(N)
        times.append(time.perf_counter() - start)
    return sum(times) / repetitions  # 평균 실행 시간 반환
    
# 테스트할 N 값 생성 (1부터 1,000,000까지 100개 간격)
N_values = np.linspace(1, 2_000, 100, dtype=int)

numpy_times = []
random_times = []
random_randint_times = []
# TODO 각 함수의 실행 시간 계산 위 배열에 추가
for N in N_values:
    numpy_times.append(measure_time(gen_r_array_choices_numpy, N))
    random_times.append(measure_time(gen_r_array_choices, N))
    random_randint_times.append(measure_time(gen_r_array_choices_numpy_rdi, N))
    
# 선그래프 그리기
plt.figure(figsize=(10, 6)) # 차트 크기 -> 그래프의 가로 크기 10인치, 세로 크기 6인치
plt.xlabel('Array Size') # x축 레이블
plt.ylabel('Time') # y축 레이블
plt.title('Performance : numpy VS random VS randint')
plt.grid(True) # 그래프에 그리드 ON
plt.plot(N_values, numpy_times, label='numpy', color='blue')
plt.plot(N_values, random_times, label='random', color='red')
plt.plot(N_values, random_randint_times, label= 'randint', color='yellow') 
plt.legend() # 범례를 추가
plt.show()
```
